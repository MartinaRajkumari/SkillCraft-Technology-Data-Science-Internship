```python
#INDIA POPULATION BY GENDER(2000-2024)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("world bank dataset.csv")
categories = ["Population, total", "Population, female", "Population, male"]
filtered_df = df[df["Series Name"].isin(categories)]

years = list(range(2000, 2025))
year_columns = [col for col in df.columns if "[YR" in col]

pop_total = filtered_df[filtered_df["Series Name"] == "Population, total"][year_columns].values.flatten().astype(float)
pop_female = filtered_df[filtered_df["Series Name"] == "Population, female"][year_columns].values.flatten().astype(float)
pop_male   = filtered_df[filtered_df["Series Name"] == "Population, male"][year_columns].values.flatten().astype(float)

bar_width = 0.25
x = range(len(years))
plt.figure(figsize=(16, 8))
plt.bar([i - bar_width for i in x], pop_total, width=bar_width, color="lightgreen")
plt.bar(x, pop_female, width=bar_width, color="pink")
plt.bar([i + bar_width for i in x], pop_male, width=bar_width, color="lightblue")
plt.xlabel("Year", fontsize=25)
plt.ylabel("Population", fontsize=25)
plt.title(" Population of India by Gender (20002024)", fontsize=25)
plt.xticks(ticks=x, labels=years, rotation=45)
plt.legend(["Total", "Female", "Male"])
plt.tight_layout()
plt.show()
```

```python
#INDIA POPULATION BY AGE GROUP AND GENDER(2000-2024)

age_groups = {
    "014": {
        "female": "Population ages 0-14, female",
        "male": "Population ages 0-14, male"
    },
    "1564": {
        "female": "Population ages 15-64, female",
        "male": "Population ages 15-64, male"
    },
    "65+": {
        "female": "Population ages 65 and above, female",
        "male": "Population ages 65 and above, male"
    }
}
year_columns = [col for col in df.columns if "[YR" in col]

years = list(range(2000, 2025))
x = np.arange(len(years))
bar_width = 0.27

female_0_14 = df[df["Series Name"] == age_groups["014"]["female"]][year_columns].values.flatten().astype(float)
male_0_14   = df[df["Series Name"] == age_groups["014"]["male"]][year_columns].values.flatten().astype(float) 
female_15_64 = df[df["Series Name"] == age_groups["1564"]["female"]][year_columns].values.flatten().astype(float)
male_15_64   = df[df["Series Name"] == age_groups["1564"]["male"]][year_columns].values.flatten().astype(float)
female_65 = df[df["Series Name"] == age_groups["65+"]["female"]][year_columns].values.flatten().astype(float)
male_65   = df[df["Series Name"] == age_groups["65+"]["male"]][year_columns].values.flatten().astype(float)
plt.figure(figsize=(20, 8))

plt.bar(x - bar_width, female_0_14, width=bar_width, color="pink", edgecolor="black")
plt.bar(x - bar_width, male_0_14, width=bar_width, bottom=female_0_14, color="lightblue", edgecolor="black", label="Age 014")
plt.bar(x, female_15_64, width=bar_width, color="pink", edgecolor="black")
plt.bar(x, male_15_64, width=bar_width, bottom=female_15_64, color="lightblue", edgecolor="black", label="Age 1564")
plt.bar(x + bar_width, female_65, width=bar_width, color="pink", edgecolor="black")
plt.bar(x + bar_width, male_65, width=bar_width, bottom=female_65, color="lightblue", edgecolor="black", label="Age 65+")

for i, year in enumerate(years):
    plt.text(x[i] - bar_width, 0, "014", ha='center', va='top', fontsize=8, rotation=90)
    plt.text(x[i], 0, "1564", ha='center', va='top', fontsize=8, rotation=90)
    plt.text(x[i] + bar_width, 0, "65+", ha='center', va='top', fontsize=8, rotation=90)

plt.xlabel("Years", labelpad=22, fontsize=25)
plt.ylabel("Population", fontsize=25)
plt.title("Population of India by Age Group and Gender (20002024)", fontsize=25)
plt.xticks(x, years, rotation=45)
plt.tick_params(axis='x', pad=19)  
plt.grid(axis='y')
plt.legend(["Female", "Male"])
plt.tight_layout()
plt.show()
```
