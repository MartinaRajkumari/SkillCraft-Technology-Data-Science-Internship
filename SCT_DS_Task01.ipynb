```python
#INDIA POPULATION BY GENDER(2000-2024)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("world bank dataset.csv")
categories = ["Population, total", "Population, female", "Population, male"]
filtered_df = df[df["Series Name"].isin(categories)]

year_columns = [f"{year} [YR{year}]" for year in range(2000, 2025)]
years = list(range(2000, 2025))

pop_total = filtered_df[filtered_df["Series Name"] == "Population, total"][year_columns].values.flatten().astype(float)
pop_female = filtered_df[filtered_df["Series Name"] == "Population, female"][year_columns].values.flatten().astype(float)
pop_male   = filtered_df[filtered_df["Series Name"] == "Population, male"][year_columns].values.flatten().astype(float)

bar_width = 0.25
x = range(len(years))
plt.figure(figsize=(16, 8))
plt.bar([i - bar_width for i in x], pop_total, width=bar_width, color="lightgreen", label="Total")
plt.bar(x, pop_female, width=bar_width, color="pink", label="Female")
plt.bar([i + bar_width for i in x], pop_male, width=bar_width, color="lightblue", label="Male")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("India Population by Gender (2000â€“2024)")
plt.xticks(ticks=x, labels=years, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#INDIA POPULATION BY AGE GROUP AND GENDER(2000-2024)

age_groups = {
    "0â€“14": {
        "female": "Population ages 0-14, female",
        "male": "Population ages 0-14, male"
    },
    "15â€“64": {
        "female": "Population ages 15-64, female",
        "male": "Population ages 15-64, male"
    },
    "65+": {
        "female": "Population ages 65 and above, female",
        "male": "Population ages 65 and above, male"
    }
}

year_columns = [f"{year} [YR{year}]" for year in range(2000, 2025)]
years = list(range(2000, 2025))
x = np.arange(len(years))
bar_width = 0.25

female_0_14 = df[df["Series Name"] == age_groups["0â€“14"]["female"]][year_columns].values.flatten().astype(float)
male_0_14   = df[df["Series Name"] == age_groups["0â€“14"]["male"]][year_columns].values.flatten().astype(float)
female_15_64 = df[df["Series Name"] == age_groups["15â€“64"]["female"]][year_columns].values.flatten().astype(float)
male_15_64   = df[df["Series Name"] == age_groups["15â€“64"]["male"]][year_columns].values.flatten().astype(float)
female_65 = df[df["Series Name"] == age_groups["65+"]["female"]][year_columns].values.flatten().astype(float)
male_65   = df[df["Series Name"] == age_groups["65+"]["male"]][year_columns].values.flatten().astype(float)
plt.figure(figsize=(16, 8))

plt.bar(x - bar_width, female_0_14, width=bar_width, color="pink", edgecolor="black")
plt.bar(x - bar_width, male_0_14, width=bar_width, bottom=female_0_14, color="lightblue", edgecolor="black", label="Age 0â€“14")
plt.bar(x, female_15_64, width=bar_width, color="pink", edgecolor="black")
plt.bar(x, male_15_64, width=bar_width, bottom=female_15_64, color="lightblue", edgecolor="black", label="Age 15â€“64")
plt.bar(x + bar_width, female_65, width=bar_width, color="pink", edgecolor="black")
plt.bar(x + bar_width, male_65, width=bar_width, bottom=female_65, color="lightblue", edgecolor="black", label="Age 65+")

for i, year in enumerate(years):
    plt.text(x[i] - bar_width, 0, "0â€“14", ha='center', va='top', fontsize=8, rotation=90)
    plt.text(x[i], 0, "15â€“64", ha='center', va='top', fontsize=8, rotation=90)
    plt.text(x[i] + bar_width, 0, "65+", ha='center', va='top', fontsize=8, rotation=90)

plt.xlabel("Years", labelpad=22)
plt.ylabel("Population")
plt.title("India Population by Age Group and Gender (2000â€“2024)")
plt.xticks(x, years, rotation=45)
plt.tick_params(axis='x', pad=19)  # Moves tick labels (years) downward
plt.grid(axis='y')
handles = [
    plt.Rectangle((0, 0), 1, 1, color="pink", edgecolor="black", label="Female"),
    plt.Rectangle((0, 0), 1, 1, color="lightblue", edgecolor="black", label="Male")
]
plt.legend(handles=handles, loc="upper left")
plt.tight_layout()
plt.show()
```
